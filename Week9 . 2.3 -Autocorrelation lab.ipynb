```python
import pandas as pd
import seaborn as sb
import matplotlib.pylab as plt 
%matplotlib inline
```

```python
companies= ['aapl','ibm','intc','msft','orcl']
```

```python
df = pd.DataFrame()
for code in companies:
    data = pd.read_csv(code +'.csv', na_values=['-'])
    data['Code'] = code
    df = df.append(data)
```

```python
df.dropna(axis=0,inplace=True)
```

```python
print len(df)
df.head()
```

```python
df.columns.values
```

```python
df['Date'] = pd.to_datetime(df['\xef\xbb\xbfDate'])
df.set_index('Date', inplace=True)

df.drop('\xef\xbb\xbfDate', axis=1, inplace=True )

df['Year'] = df.index.year
df['Month'] = df.index.month
```

```python
df.head()
```

```python
sb.factorplot(
    x='Month',
    y='Volume',
#     hue='Code',
    data=df,
    kind='box'
)
```

```python
plt.figure(figsize=(10,8))
for k, g in df.groupby(['Code']):
    plt.plot(g['Close'], label = k)
plt.legend(loc='best')    
plt.show()
```

```python
plt.figure(figsize=(10,8))
for k, g in df.groupby(['Code']):
    plt.plot(g['Volume'], label = k)
plt.legend(loc='best')    
plt.show()
```

```python
def corre_by_day(comp):
    print "One day Mean: ",comp.resample('D').mean().autocorr(lag=1)
    print "One month Mean: ",comp.resample('D').mean().autocorr(lag=30)
    print "Two months Mean: ",comp.resample('D').mean().autocorr(lag=60)
    print ""
```

```python
print "\nCorreleations of closing price for all companies by mean\n"
for k, g in df.groupby(['Code']):
    print k
    corre_by_day(g['Close'])
```

```python
print "\nCorreleations of Volume for all companies by mean\n"
for k, g in df.groupby(['Code']):
    corre_by_day(g['Volume'])
```
